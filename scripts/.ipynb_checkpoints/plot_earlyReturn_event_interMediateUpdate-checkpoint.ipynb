```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fmu_dir = '/home/nyoshimatsu/work/fmi/Reference-FMUs'
temp_dir = 'build/temp'
```

```python
def plot_csv(csv_file, title):
    
    csv_file_path = fmu_dir + '/' + temp_dir + '/' + csv_file
    df = pd.read_csv(csv_file_path, index_col=0)

    df.plot(subplots=True, layout=(1, 2), figsize=(14,4), grid=True, title=title, style=['b.-'])
    
    # ax1.set_ylabel('[m]')
    # ax2.set_ylabel('[m/s]')
    
    # plot.show()
    
    return
```

以下の時間を設定。
・ Solve Step時間：0.01[秒]
・ Communication Step時間：0.1[秒]

```python
plot_csv('BouncingBall_cs_out.csv', 'BouncingBall_cs_out')
```

```python
plot_csv('cs_early_return_out.csv', 'cs_early_return')
```
